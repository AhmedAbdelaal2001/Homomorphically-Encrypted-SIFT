```python
import os
from PIL import Image
import matplotlib.pyplot as plt
import pickle
import cv2

# Specify the folder name
folder_name = 'data'

# Check if the folder exists, if not, create it
if not os.path.exists(folder_name):
    os.makedirs(folder_name)

# Define the path to save the image
save_path = os.path.join(folder_name, 'image4.jpg')

# Load and show the image if it exists
if os.path.isfile(save_path):
    loaded_image = Image.open(save_path)
    plt.imshow(loaded_image)
    plt.show()
else:
    print(f"No image found at {save_path}")


# Define the path to load the keypoints
keypoints_path = os.path.join(folder_name, 'image4_keypoints.pkl')
# Define the path to load the descriptors
descriptors_path = os.path.join(folder_name, 'image4_descriptors.pkl')

# Load keypoints
with open(keypoints_path, 'rb') as f:
    keypoints_serializable = pickle.load(f)
loaded_keypoints = [cv2.KeyPoint(x=p[0], y=p[1], size=p[2], angle=p[3], response=p[4], octave=p[5], class_id=p[6]) for p in keypoints_serializable]

# Load descriptors
with open(descriptors_path, 'rb') as f:
    loaded_descriptors = pickle.load(f)
```

```python
bf = cv2.BFMatcher()
matches = bf.knnMatch(loaded_descriptors, loaded_descriptors, k=2)
good_matches = []
ratio_threshold = 0.75  # Commonly used threshold
for m, n in matches:
    if m.distance < ratio_threshold * n.distance:
        good_matches.append(m)
matches = sorted(good_matches, key = lambda x:x.distance)
len(matches)
```

```python
loaded_image_gray = (rgb2gray(loaded_image) * 255).astype(np.int64)

matching_result = cv2.drawMatches(encryptImage(loaded_image_gray).astype(np.uint8), loaded_keypoints, encryptImage(loaded_image_gray).astype(np.uint8), loaded_keypoints, matches, None)
# Create a figure with a white background
fig, ax = plt.subplots()
fig.patch.set_facecolor('white')

# Display the image without axes
ax.imshow(matching_result)
ax.axis('off')  # Hide the axes

plt.show()
```
